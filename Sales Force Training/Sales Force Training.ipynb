# Import Libraries

```python
import pandas as pd
import statistics
```

```python
# Create DataFrame
data = pd.DataFrame({'TransactionsAmount': [100, 150, 50, 100, 130, 120, 100, 85, 70, 150, 150, 120, 50, 100, 100, 140, 90, 150, 50, 90, 120, 100, 110, 75, 65]})
data
```

```python
# count length of data
length = len(data)
print("Length of data:", length)
```

```python
# Central Tendency
mean = data['TransactionsAmount'].mean()
median = data['TransactionsAmount'].median()
mode = statistics.mode(data['TransactionsAmount'])
```

```python
#Tulis analisismu dari Central Tendency di sini
print("The average value (mean) in the Transaction Amount column is", mean)
print("The median value in the Transaction Amount column is", median)
print("The most frequent value (mode) in the Transaction Amount column", mode)
```

```python
import numpy as np

# Variability
variance = np.var(data)
standard_deviation = np.std(data)
maximum = np.max(data)
minimum = np.min(data)
quartile_1 = np.quantile(data, 0.25)
quartile_2 = np.quantile(data, 0.5)
quartile_3 = np.quantile(data, 0.75)
```

```python
# Analysis of Variability
print("The variance of Transaction Amount is", variance)
print("The standard deviation of Transaction Amount is {:.3f}".format(float(standard_deviation)))
print("The maximum value of Transaction Amount is", maximum)
print("The minimum value of Transaction Amount is", minimum)
print("The first quartile (Q1) of Transaction Amount is", quartile_1)
print("The second quartile (median) of Transaction Amount is", quartile_2)
print("The third quartile (Q3) of Transaction Amount is", quartile_3)
```

Define H0 and H1

H0: The average of the 25 customer transactions amount = 100

H1: The average of the 25 customer transactions amount more than 100

Define alpha = 0.05

```python
# Perform one-sample t-test
t_statistic, p_value = stats.ttest_1samp(a=data['TransactionsAmount'], popmean=100)

# Round the result to three decimal places
t_statistic = round(t_statistic, 3)
p_value = round(p_value, 3)

# Print the t-statistic and p-value
print("t-statistic:", t_statistic)
print("p-value:", p_value)
```

Conclude with alpha and p-value

Since the P-value is greater than alpha (0.05), we fail to reject H0.

Conclusion: Sales force training is effective, on average out of 25 customers the number of transactions is more than 100

```python
import scipy.stats as stats

# Calculate the t-value for a one-tailed test
t_value = stats.t.ppf(1 - 0.05, 24)

# Round the result to three decimal places
t_value = round(t_value, 3)

# Print the t-value
print("t-value:", t_value)
```

Critical region result: t < 1.73

Since the t-statistic is not in the critical region, we fail to reject H0

Conclusion: Sales force training is effective, on average out of 25 customers the number is more than 100
